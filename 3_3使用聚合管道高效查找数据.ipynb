```python
import charts
import pymongo
```

```python
client = pymongo.MongoClient('localhost', 27017)
ceshi = client['ganji']
item_info = ceshi['item_info']
```

```python
for i in item_info.find().limit(200):
    print(i)
```

```python
pipeline = [
    {'$match':{'$and':[{'pub_date':'2015.12.24'},{'area':['朝阳']}]}}
]
for i in item_info.aggregate(pipeline):
    print(i)
```

```python
pipeline = [
    {'$match':{'pub_date':'2015.12.24'}},
    {'$group':{'_id':'$area','counts':{'$sum':1}}}
]
for i in item_info.aggregate(pipeline):
    print(i)
```

```python
pipeline = [
    {'$match':{'pub_date':'2015.12.24'}},
    {'$group':{'_id':'$area','counts':{'$sum':1}}},
    {'$sort':{'counts':-1}}
]
for i in item_info.aggregate(pipeline):
    print(i)
```

```python
pipeline = [
    {'$match':{'pub_date':'2015.12.24'}},
    {'$group':{'_id':'$area','counts':{'$sum':1}}},
    {'$sort':{'counts':-1}},
    {'$limit':5}
]
for i in item_info.aggregate(pipeline):
    print(i)
```

```python
for i in item_info.find({},{'cates':1,'_id':0}).limit(200):
    print(i)
```

```python
#统计某一天二手物品的类目分布占比
pipeline2 = [
    {'$match':{'pub_date':'2015.12.24'}},
    {'$group':{'_id':{'$slice':['$cates',2,1]},'counts':{'$sum':1}}},
    {'$sort':{'counts':-1}},
]
for i in item_info.aggregate(pipeline2):
    print(i)
```

```python
def data_gen(date):
    pipeline = [
        {'$match':{'pub_date':date}},
        {'$group':{'_id':{'$slice':['$cates',2,1]},'counts':{'$sum':1}}},
        {'$sort':{'counts':-1}},
    ]
    for i in item_info.aggregate(pipeline):
        yield[i['_id'][0],i['counts']]
```

```python
options = {
    'chart'   : {'zoomType':'xy'},
    'title'   : {'text': '发帖量统计'},
    'subtitle': {'text': '可视化统计图表'},
    }

series =  [{
    'type': 'pie',
    'name': 'pie charts',
    'data':[i for i in data_gen('2015.12.24')]}]

charts.plot(series,options=options,show='inline')
```
